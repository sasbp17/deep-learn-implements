## Activation Functions

---

1. Sigmoid function
2. TanH function
3. ReLU function
4. Leaky ReLU function
5. ELU function

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. Sigmoid Function

```python
def sigmoid_func(z):
    return 1 / (1 + np.exp(-z))

fig = plt.figure()
ax = plt.axes()

x = np.linspace(-5, 5, num=1000)
y = sigmoid_func(x)
plt.plot(x, y)
plt.title("Sigmoid Activation Function")
plt.xlabel("Input - x")
plt.ylabel("Output - Sigmoid(x)")
plt.show()
```

#### 2. tanh Function

```python
def tanh_func(z):
    return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))

fig = plt.figure()
ax = plt.axes()

x = np.linspace(-3, 3, num=1000)
y = tanh_func(x)
plt.plot(x, y)
plt.title("tanh Activation Function")
plt.xlabel("Input - x")
plt.ylabel("Output - tanh(x)")
plt.show()
```

#### 3. ReLU Function

```python
def relu_func(z):
    return np.maximum(0, z)

fig = plt.figure()
ax = plt.axes()

x = np.linspace(-5, 10)
y = [relu_func(i) for i in x] 
plt.plot(x, y)
plt.title("ReLU Activation Function")
plt.xlabel("Input - x")
plt.ylabel("Output - ReLu(x)")
plt.show()
```

#### 4. Leaky ReLU Function

```python
def lrelu_func(z, a):
    # a is the small slope for negative inputs
    y = []
    for i in z:
        if i < 0:
            y.append(a * i)
        else:
            y.append(i)
    return y

fig = plt.figure()
ax = plt.axes()

x = np.linspace(-2, 0.8, num=1000)
y = lrelu_func(x, 0.3) # slope increased steeply for demo purposes
plt.plot(x, y)
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input - x")
plt.ylabel("Output - Leaky ReLU(x)")
plt.show()
```

#### 5. ELU Function

```python
def elu_func(z, a):
    # a is the small slope for negative inputs
    y = []
    for i in z:
        if i < 0:
            y.append(a * (np.exp(i) - 1))
        else:
            y.append(i)
    return y

fig = plt.figure()
ax = plt.axes()

x = np.linspace(-2, 0.8, num=1000)
y = elu_func(x, 0.3) # slope increased steeply for demo purposes
plt.plot(x, y)
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input - x")
plt.ylabel("Output - Leaky ReLU(x)")
plt.show()
```
